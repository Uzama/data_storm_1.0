```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
train_data = pd.read_csv("../data/credit_card_default_train.csv")
test_data = pd.read_csv("../data/credit_card_default_test.csv")
```

```python
train_data.drop('Client_ID',axis=1,inplace=True)
```

```python
train_data.columns = train_data.columns.map(str.lower)
```

```python
train_data.columns
```

```python
col_to_norm = ['balance_limit_v1','age', 'due_amt_july', 'due_amt_aug', 'due_amt_sep', 'due_amt_oct',
       'due_amt_nov', 'due_amt_dec', 'paid_amt_july', 'paid_amt_aug',
       'paid_amt_sep', 'paid_amt_oct', 'paid_amt_nov', 'paid_amt_dec']
train_data[col_to_norm] = train_data[col_to_norm].apply(lambda x : (x-np.mean(x))/np.std(x))
```
